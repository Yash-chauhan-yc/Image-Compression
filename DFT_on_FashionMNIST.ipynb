```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img0 = cv2.imread('/content/00000.jpg',0)
img2 = cv2.imread('/content/00003.jpg',0)
img1 = cv2.imread('/content/00005.jpg',0)
img3 = cv2.imread('/content/00001.jpg',0)
img4 = cv2.imread('/content/00002.jpg',0)
img5 = cv2.imread('/content/00004.jpg',0)
img6 = cv2.imread('/content/00006.jpg',0)
img7 = cv2.imread('/content/00007.jpg',0)
img8 = cv2.imread('/content/00008.jpg',0)
img9 = cv2.imread('/content/00009.jpg',0)

img0_float32 = np.float32(img0)
img1_float32 = np.float32(img1)
img2_float32 = np.float32(img2)
img3_float32 = np.float32(img3)
img4_float32 = np.float32(img4)
img5_float32 = np.float32(img5)
img6_float32 = np.float32(img6)
img7_float32 = np.float32(img7)
img8_float32 = np.float32(img8)
img9_float32 = np.float32(img9)

dft0 = cv2.dft(img0_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft1 = cv2.dft(img1_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft2 = cv2.dft(img2_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft3 = cv2.dft(img3_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft4 = cv2.dft(img4_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft5 = cv2.dft(img5_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft6 = cv2.dft(img6_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft7 = cv2.dft(img7_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft8 = cv2.dft(img8_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft9 = cv2.dft(img9_float32, flags = cv2.DFT_COMPLEX_OUTPUT)

dft_shift0 = np.fft.fftshift(dft0)
dft_shift1 = np.fft.fftshift(dft1)
dft_shift2 = np.fft.fftshift(dft2)
dft_shift3 = np.fft.fftshift(dft3)
dft_shift4 = np.fft.fftshift(dft4)
dft_shift5 = np.fft.fftshift(dft5)
dft_shift6 = np.fft.fftshift(dft6)
dft_shift7 = np.fft.fftshift(dft7)
dft_shift8 = np.fft.fftshift(dft8)
dft_shift9 = np.fft.fftshift(dft9)

magnitude_spectrum0 = 20*np.log(cv2.magnitude(dft_shift0[:,:,0],dft_shift0[:,:,1]))
magnitude_spectrum1 = 20*np.log(cv2.magnitude(dft_shift1[:,:,0],dft_shift1[:,:,1]))
magnitude_spectrum2 = 20*np.log(cv2.magnitude(dft_shift2[:,:,0],dft_shift2[:,:,1]))
magnitude_spectrum3 = 20*np.log(cv2.magnitude(dft_shift3[:,:,0],dft_shift3[:,:,1]))
magnitude_spectrum4 = 20*np.log(cv2.magnitude(dft_shift4[:,:,0],dft_shift4[:,:,1]))
magnitude_spectrum5 = 20*np.log(cv2.magnitude(dft_shift5[:,:,0],dft_shift5[:,:,1]))
magnitude_spectrum6 = 20*np.log(cv2.magnitude(dft_shift6[:,:,0],dft_shift6[:,:,1]))
magnitude_spectrum7 = 20*np.log(cv2.magnitude(dft_shift7[:,:,0],dft_shift7[:,:,1]))
magnitude_spectrum8 = 20*np.log(cv2.magnitude(dft_shift8[:,:,0],dft_shift8[:,:,1]))
magnitude_spectrum9 = 20*np.log(cv2.magnitude(dft_shift9[:,:,0],dft_shift9[:,:,1]))


plt.subplot(121),plt.imshow(img0, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum0, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()         

plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum1, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img2, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum2, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img3, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum3, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img4, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum4, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img5, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum5, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img6, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum6, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img7, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum7, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img8, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum8, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   

plt.subplot(121),plt.imshow(img9, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum9, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()   
```
